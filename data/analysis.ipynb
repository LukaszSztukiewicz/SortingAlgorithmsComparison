```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_folder = "./out/"
```

```python
def normalize(arr):
    for i in range(len(arr)):
        arr[i]*=1.34
    return arr
```

```python
#Speed comparison
df = pd.read_csv(f"{data_folder}randomArray_speed_comparison.csv")
for col in df.columns.values[1:]:
    plt.plot(df.iloc[:, 0].values, normalize(df[col].values), label=col)
plt.xlim(1, df.iloc[:, 0].values[-1])
plt.xlabel("Number of elements")
plt.ylabel("Time in seconds")
plt.legend(loc='best')
plt.savefig(f"speed_comparison.pdf", dpi=600)
```
